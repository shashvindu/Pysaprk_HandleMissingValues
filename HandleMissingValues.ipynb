```
import random
import pandas as pd
from pyspark.sql import SparkSession
from pyspark.sql import functions as F



# Create a Pandas dataframe with 10 columns and 100 rows
data = {
    "col_1": [random.randint(1, 100) for i in range(100)],
    "col_2": [random.randint(1, 100) for i in range(100)],
    "col_3": [random.randint(1, 100) for i in range(100)],
    "col_4": [random.randint(1, 100) for i in range(100)],
    "col_5": [random.randint(1, 100) for i in range(100)],
    "col_6": [random.randint(1, 100) for i in range(100)],
    "col_7": [random.randint(1, 100) for i in range(100)],
    "col_8": [random.randint(1, 100) for i in range(100)],
    "col_9": [random.randint(1, 100) for i in range(100)],
    "col_10": [random.randint(1, 100) for i in range(100)],
}
```

```
df = pd.DataFrame(data)


df
```

```
# Replace some values with missing or null values
df.iloc[:, 0:5] = df.iloc[:, 0:5].mask(pd.Series(data=[True if random.random() < 0.2 else False for i in range(100)]))

df
```

```
# Convert the Pandas dataframe to a Spark dataframe
spark_df = spark.createDataFrame(df)

display(spark_df)
```

```
# Show the number of missing or null values in each column
print("Number of missing or null values in each column:")
spark_df.select([F.count(F.when(F.isnull(c), c)).alias(c) for c in spark_df.columns]).show()
```

```

# Replace missing or null values with the mean of the column
spark_df = spark_df.na.fill(spark_df.agg(F.mean(spark_df["col_1"])).first()[0], subset=["col_1"])

```

```
# Show the number of missing or null values in each column after handling
print("Number of missing or null values in each column after handling:")
spark_df.select([F.count(F.when(F.isnull(c), c)).alias(c) for c in spark_df.columns]).show()

```

```
import pandas as pd
from pyspark.sql import SparkSession
from pyspark.sql import functions as F



# File location and type
file_location = "/FileStore/tables/covid_worldwide.csv"
file_type = "csv"

# CSV options
infer_schema = "True"
first_row_is_header = "True"
delimiter = ","

# The applied options are for CSV files. For other file types, these will be ignored.
spark_df = spark.read.format(file_type) \
  .option("inferSchema", infer_schema) \
  .option("header", first_row_is_header) \
  .option("sep", delimiter) \
  .load(file_location)

display(spark_df)
```

```
# Show the number of missing or null values in each column
print("Number of missing or null values in each column before handling:")
display(spark_df.select([F.count(F.when(F.isnull(c), c)).alias(c) for c in spark_df.columns]))

# Replace missing or null values with the mean of the column
mean_values = spark_df.agg(*[F.mean(c).alias(c) for c in spark_df.columns if c in spark_df.columns]).toPandas().to_dict('records')[0]
spark_df = spark_df.na.fill(mean_values)

# Show the number of missing or null values in each column after replacing with mean
print("Number of missing or null values in each column after replacing with mean:")
display(spark_df.select([F.count(F.when(F.isnull(c), c)).alias(c) for c in spark_df.columns]))

```

```
# Replace missing or null values with the median of the column
median_values = spark_df.agg(*[F.percentile_approx(c, 0.5).alias(c) for c in spark_df.columns if c in spark_df.columns]).toPandas().to_dict('records')[0]
spark_df = spark_df.na.fill(median_values)

# Show the number of missing or null values in each column after replacing with median
print("Number of missing or null values in each column after replacing with median:")
display(spark_df.select([F.count(F.when(F.isnull(c), c)).alias(c) for c in spark_df.columns]))

```

```
# Replace missing or null values with the mode of the column
mode_values = spark_df.groupBy(*[c for c in spark_df.columns if c in spark_df.columns]).agg(*[F.first(c).alias(c) for c in spark_df.columns if c in spark_df.columns]).first().asDict()
spark_df = spark_df.na.fill(mode_values)

# Show the number of missing or null values in each column after replacing with mode
print("Number of missing or null values in each column after replacing with mode:")
spark_df.select([F.count(F.when(F.isnull(c), c)).alias(c) for c in spark_df.columns]).show()
display(spark_df)
```

```
# Drop missing or null values
spark_df = spark_df.dropna()

# Show
print("Number of missing or null values in each column after dropping missing or null values:")
spark_df.select([F.count(F.when(F.isnull(c), c)).alias(c) for c in spark_df.columns]).show()
display(spark_df)
```
